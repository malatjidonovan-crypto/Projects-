```python
from collections import defaultdict

# Dataset
data = [
    ("Yes", "No", "No"),
    ("Yes", "No", "No"),
    ("Yes", "Yes", "Yes"),
    ("No", "Yes", "Yes"),
    ("No", "Yes", "Yes")
]

# Count
class_counts = defaultdict(int)
feature_counts = defaultdict(lambda: defaultdict(int))

for age, job, label in data:
    class_counts[label] += 1
    feature_counts[label][("age", age)] += 1
    feature_counts[label][("job", job)] += 1

total = len(data)

# Prediction function
def predict(age, job):
    probs = {}
    
    for c in class_counts:
        # Prior
        prob = class_counts[c] / total
        
        # Likelihoods
        prob *= feature_counts[c][("age", age)] / class_counts[c]
        prob *= feature_counts[c][("job", job)] / class_counts[c]
        
        probs[c] = prob
    
    return max(probs, key=probs.get), probs

# Test
prediction, probs = predict("Yes", "Yes")

print("Prediction:", prediction)
print("Probabilities:", probs)
```
