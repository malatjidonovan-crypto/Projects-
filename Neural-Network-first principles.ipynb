```python
import numpy as np

# Sigmoid and derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Initialize weights
w1, b1 = np.random.randn(), np.random.randn()
w2, b2 = np.random.randn(), np.random.randn()

# Data
x = 0.5
y = 1

lr = 0.1

for epoch in range(1000):
    
    # ---- Forward ----
    z1 = w1 * x + b1
    a1 = sigmoid(z1)
    
    z2 = w2 * a1 + b2
    y_hat = sigmoid(z2)
    
    # ---- Loss ----
    loss = 0.5 * (y - y_hat)**2
    
    # ---- Backward ----
    dL_dyhat = y_hat - y
    dyhat_dz2 = sigmoid_derivative(y_hat)
    
    delta2 = dL_dyhat * dyhat_dz2
    
    dL_dw2 = delta2 * a1
    dL_db2 = delta2
    
    delta1 = delta2 * w2 * sigmoid_derivative(a1)
    
    dL_dw1 = delta1 * x
    dL_db1 = delta1
    
    # ---- Update ----
    w2 -= lr * dL_dw2
    b2 -= lr * dL_db2
    
    w1 -= lr * dL_dw1
    b1 -= lr * dL_db1

print("Final prediction:", y_hat)
```
