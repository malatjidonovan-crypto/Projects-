```python
import math

def entropy(labels):
    total = len(labels)
    counts = {}
    
    for label in labels:
        counts[label] = counts.get(label, 0) + 1
    
    H = 0
    for count in counts.values():
        p = count / total
        H -= p * math.log2(p)
    
    return H

# Dataset
data = [(1,0),(2,0),(3,1),(4,1),(5,1)]

# Split function
def split(data, threshold):
    left = [y for x,y in data if x <= threshold]
    right = [y for x,y in data if x > threshold]
    return left, right

# Test split x > 3
left, right = split(data, 2)

H_total = entropy([y for _,y in data])
H_left = entropy(left)
H_right = entropy(right)

H_split = (len(left)/len(data))*H_left + (len(right)/len(data))*H_right

IG = H_total - H_split

print("Entropy before:", H_total)
print("Entropy after:", H_split)
print("Information Gain:", IG)
```
