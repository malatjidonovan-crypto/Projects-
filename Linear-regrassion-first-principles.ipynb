```python
import numpy as np

# Data
X = np.array([1, 2, 3,4,6])
Y = np.array([2, 4, 6,8,12])

# Initialize parameters
w = 0.0
b = 0.0

lr = 0.01
epochs = 1000
n = len(X)

for epoch in range(epochs):
    
    # Predictions
    Y_pred = w * X + b
    
    # Compute gradients
    dw = (-2/n) * np.sum(X * (Y - Y_pred))
    db = (-2/n) * np.sum(Y - Y_pred)
    
    # Update
    w -= lr * dw
    b -= lr * db

print("Final w:", w)
print("Final b:", b)
```
